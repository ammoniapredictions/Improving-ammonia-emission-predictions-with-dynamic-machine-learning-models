# Model

**We do not use the 'AmmoniaRNN' module from ./functions/rnn_module.py. We need to modify it to get the latent variables in the ouput.**

```python
class AmmoniaRNN(nn.Module):
    
    def __init__(self, 
                 input_size, 
                 output_size, 
                 hidden_size, 
                 num_layers,
                 nonlinearity, 
                 bidirectional,
                 mlp,
                 with_embeddings = False, cat_dims = None, embedding_dims = None):
        
        super(AmmoniaRNN, self).__init__()
        self.input_size = input_size
        self.output_size = output_size
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.nonlinearity = nonlinearity
        self.bidirectional = bidirectional
        self.mlp = mlp
        self.with_embeddings = with_embeddings

        D = 1 + 1 * bidirectional
        
        if (self.with_embeddings):
       
            # Create an embedding layer for each categorical variable
            self.embeddings = nn.ModuleList([
                nn.Embedding(num_embeddings = cat_dim, embedding_dim = embed_dim)
                for cat_dim, embed_dim in zip(cat_dims, embedding_dims)
            ])
            
            input_size = input_size - len(cat_dims) + sum(embedding_dims)           
        
        # RNN layer
        self.rnn = nn.RNN(input_size, 
                          hidden_size, 
                          num_layers = num_layers,
                          nonlinearity = nonlinearity, 
                          bidirectional = bidirectional)    
        
        # Fully connected layer(s)
        if self.mlp:
            self.fc1 = nn.Linear(hidden_size * D, 6)
            self.relu = nn.ReLU()
            self.fc2 = nn.Linear(6, output_size)
        else:
            self.fc = nn.Linear(hidden_size * D, output_size)

    def forward(self, x):

        if (self.with_embeddings):

            x_continuous = x[0]
            x_categoricals = x[1]
            
            # Get embeddings for each categorical variable
            x_embeds = [embed(x_cat) for embed, x_cat in zip(self.embeddings, x_categoricals)]
            
            # Concatenate all embeddings with the continuous features
            x = torch.cat([x_continuous] + x_embeds, dim = -1)
        
        # Forward pass through RNN
        h1, _ = self.rnn(x)

        if self.mlp:
            h2 = self.fc1(h1)
            h3 = self.relu(h2)
            out = self.fc2(h3)
        else:
            out = self.fc(h1)
        
        return [out, x_embeds]
```

# Predictions on one sample

```python
pmids_test = list_test_pmids [0]
```

```python
selected_model = "results/06_model_comparison/models/sample_0_model_9.pth"
```

```python
data_test = data[data['pmid'].isin(pmids_test)]
pmids_test = data_test['pmid'].unique()
```

```python
torch.manual_seed(1)
model = AmmoniaRNN(input_size = input_size, 
                   output_size = output_size, 
                   hidden_size = hidden_size, 
                   nonlinearity = nonlinearity,
                   num_layers = num_layers,
                   bidirectional = bidirectional,
                   mlp = mlp,
                   with_embeddings = with_embeddings, 
                   cat_dims = cat_dims, 
                   embedding_dims = embedding_dims).to(DEVICE)
```

```python
data_predictions = data_test.copy()
data_predictions['prediction_ecum'] = None
data_predictions['prediction_delta_ecum'] = None
    
with torch.no_grad():

    all_predictions = torch.empty(0).to(DEVICE)
    all_h = torch.empty(0).to(DEVICE)

    model.load_state_dict(torch.load(selected_model, weights_only = True, map_location = torch.device('cpu')))

    for i in pmids_test:

        x = mf.generate_tensors_predictors (data_predictions, i, with_embeddings, device = DEVICE)

        model_outputs = model (x)
        
        y = model_outputs[0]
        all_predictions = torch.cat ((all_predictions, y.squeeze()), 0)

        h = model_outputs[1]
        h = torch.cat (h, 1)
        all_h = torch.cat ((all_h, h.squeeze()), 0)

    data_predictions['prediction_delta_ecum'] = all_predictions.to("cpu").detach()

data_predictions["prediction_ecum"] = data_predictions.groupby("pmid")["prediction_delta_ecum"].cumsum()
data_predictions["absolute_error"] = abs(data_predictions["prediction_ecum"] - data_predictions["e.cum"])
```

# Predictions on all samples

```python
p = 15
```

```python
df_list = []

for sample in range (p):    
    
    pmids_test = list_test_pmids [sample]
        
    selected_model = f'results/06_model_comparison/models/sample_{sample}_model_9.pth'
    
    data_test = data[data['pmid'].isin(pmids_test)]
    pmids_test = data_test['pmid'].unique()
    
    torch.manual_seed(1)
    model = AmmoniaRNN(input_size = input_size, 
                       output_size = output_size, 
                       hidden_size = hidden_size, 
                       nonlinearity = nonlinearity,
                       num_layers = num_layers,
                       bidirectional = bidirectional,
                       mlp = mlp,
                       with_embeddings = with_embeddings, 
                       cat_dims = cat_dims, 
                       embedding_dims = embedding_dims).to(DEVICE)
    
    data_predictions = data_test.copy()
    data_predictions['prediction_ecum'] = None
    data_predictions['prediction_delta_ecum'] = None
        
    with torch.no_grad():
    
        all_predictions = torch.empty(0).to(DEVICE)
        all_h1 = torch.empty(0).to(DEVICE)
        all_h2 = torch.empty(0).to(DEVICE)
    
        model.load_state_dict(torch.load(selected_model, weights_only = True, map_location = torch.device('cpu')))
    
        for i in pmids_test:
    
            x = mf.generate_tensors_predictors (data_predictions, i, with_embeddings, device = DEVICE)
    
            model_outputs = model (x)
            
            y = model_outputs[0]
            all_predictions = torch.cat ((all_predictions, y.squeeze()), 0)
    
            h1 = model_outputs[1]
            all_h1 = torch.cat ((all_h1, h1.squeeze()), 0)
    
            h2 = model_outputs[2]
            all_h2 = torch.cat ((all_h2, h2.squeeze()), 0)
    
    
        data_predictions['prediction_delta_ecum'] = all_predictions.to("cpu").detach()
    
    data_predictions["prediction_ecum"] = data_predictions.groupby("pmid")["prediction_delta_ecum"].cumsum()
    data_predictions["absolute_error"] = abs(data_predictions["prediction_ecum"] - data_predictions["e.cum"])
    data_predictions["sample"] = sample
    
    data_predictions = pd.concat([data_predictions.reset_index(drop=True), pd.DataFrame (all_h1).reset_index(drop=True)], axis = 1)
    data_predictions = pd.concat([data_predictions.reset_index(drop=True), pd.DataFrame (all_h2).reset_index(drop=True)], axis = 1)
        
    df_list.append (data_predictions)
```

```python
# Création d'un DataFrame avec les résultats
tsne_df = pd.DataFrame(X_tsne_transformed, columns=['Dim1', 'Dim2'])
tsne_df["quantitative_var"] = quantitative_var  # Ajout de la variable supplémentaire

# Tracé du nuage de points coloré par la variable quantitative
plt.figure(figsize=(10, 6))
scatter = plt.scatter(tsne_df["Dim1"], tsne_df["Dim2"], 
                      c=tsne_df["quantitative_var"], cmap="coolwarm", alpha=0.75)
plt.colorbar(scatter, label="Absolute error")

plt.xlabel("t-SNE Dim 1")
plt.ylabel("t-SNE Dim 2")
plt.title("Projection des individus avec t-SNE")

plt.show()
```
