# Choice of the hidden size

```python
hidden_size = [16, 64, 128, 256, 512]
with_embeddings = [True, False]
min_mae_eval = [None]
sample = [None]
running_time = [None]

hyper_grid = pd.DataFrame(
    itertools.product(hidden_size, with_embeddings, min_mae_eval, sample, running_time), 
    columns=["hidden_size", "with_embeddings", "min_mae_eval", "sample", "running_time"]
)

hyper_grid.insert(0, "model", range(1, len(hyper_grid) + 1))

hyper_grid
```

```python
num_layers = 1
nonlinearity = "relu"
bidirectional = False
mlp = False
response = "e.cum"
```

```python
p = 15
```

```python
folder_path = "results/04_hidden_size_choice/df_res"

df_list = []

for file in os.listdir(folder_path):
    if file.endswith(".csv"): 
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path) 
        df_list.append(df)

df_res = pd.concat(df_list, ignore_index=True)
```

```python
model_index = 7

for sample in range(p): 

    with open(f'results/04_hidden_size_choice/all_mae_eval/all_mae_eval_model_{model_index}_sample_{sample}.pkl', "rb") as f:
        all_mae_eval = pickle.load(f)

    with open(f'results/04_hidden_size_choice/all_losses/all_losses_model_{model_index}_sample_{sample}.pkl', "rb") as f:
        all_losses = pickle.load(f)

    fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=False)

    axes[0].plot(all_losses[100:])
    axes[0].set_title(f'All Losses (Sample {sample})')

    axes[1].plot(all_mae_eval[100:])
    axes[1].set_title(f'All mae eval (Sample {sample})')

    plt.suptitle(f'Model {model_index} - h = {df_res["hidden_size"][model_index - 1]} - '
                 f'Embeddings = {df_res["with_embeddings"][model_index - 1]} - Sample {sample}')

    plt.tight_layout()

    plt.show()
```

```python
df_res.groupby(['model', 'hidden_size', 'with_embeddings']).size().reset_index(name='count')
```

```python
filtered_data = df_res[df_res['with_embeddings'] == True]

grouped_data = [filtered_data[filtered_data["hidden_size"] == h]["min_mae_eval"]
                for h in sorted(filtered_data["hidden_size"].unique())]

plt.figure(figsize=(10, 6))
plt.boxplot(grouped_data, labels=sorted(filtered_data["hidden_size"].unique())) 

plt.xlabel("Hidden Size")
plt.ylabel("Min MAE Eval")
plt.title("With embeddings")

plt.show()
```

```python
filtered_data = df_res[df_res['with_embeddings'] == False]

grouped_data = [filtered_data[filtered_data["hidden_size"] == h]["min_mae_eval"]
                for h in sorted(filtered_data["hidden_size"].unique())]

plt.figure(figsize=(10, 6))
plt.boxplot(grouped_data, labels=sorted(filtered_data["hidden_size"].unique())) 

plt.xlabel("Hidden Size")
plt.ylabel("Min MAE Eval")
plt.title("Without embeddings")

plt.show()
```
